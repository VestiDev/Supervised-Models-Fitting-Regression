```python
# loading libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from ggplot import *
```

```python
# define column names
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# loading training data
iris = pd.read_csv('iris.data', header=None, names=names)
iris.head()
```

```python
#summary statistics
iris.describe()
```

The iris data set was downloaded from [UC Irvine's School of Information and Computer Sciences](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/) and is on github.

## Visualizing Data

```python
# sepal width vs. sepal length
ggplot(aes(x='sepal_length', y='sepal_width', color='species'), data=iris) + geom_point()+\
    scale_color_brewer(type='qual', palette=6)
```

```python
# petal width vs. petal length
ggplot(aes(x='petal_length', y='petal_width', color="species"), data=iris) + geom_point()+\
    scale_color_brewer(type='qual', palette=6)
```
